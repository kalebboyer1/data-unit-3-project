#### Fill in blanks with a brief description of each plot type:

  
- Histogram
> used to show the distribution of numerical data by presenting the frequency of data values using bins (ranges).
- Density Plot / KDE
> Generally, the kernel density estimate is used in density plots to show the probability density function of the variable.
- Bar Chart
> used for visualizing the count of nomial/categorical data and a great alternative to the issues with pie charts.
- Line Plot
> A line plot visualizes data by connecting the data points via line segments with the measurement points in order and joined with straight line segments. 
- Univariate scatter plot
> plots different observations/values of the same variable corresponding to the index/observation number.
- Violin Plot
> similar to a box plot with the addition of a rotated kernel density plot on each side. It shows the distribution of quantitative data across several levels of one (or more) categorical variables.
- Box Plot
> used to also give descriptive statistics that are not found in histograms.
- Pie Chart
> used to visualize value distribution for nominal or ordinal data.
- Rug Plot
> consists of vertical lines at each data point. Here, the height is arbitrary. The density of the distribution can be known by how dense the tick-marks are

### Step 5 - Create Plots for Each Variable in Dataset

```python
# import seaborn
import seaborn as sns

# set standard plot type as seaborn plot
sns.set()

# import matplotlib
import matplotlib.pyplot as plt

# write command for plots to appear in jupyter cells automatically
%matplotlib inline
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

### Plot for 'age' column

```python
# create plot for age column
plt.figure(figsize=(12,6))

sns.violinplot(x=df['age'])

plt.xlabel("Age", fontsize=14)
plt.title("Density of Manager Ages")
```

##### Insights:

mid 30s to mid 60s have the majority

### Plot for 'industry' column

```python
# create plot for industry column
plt.figure(figsize=(12,6))

ax = sns.countplot(data=df, x='industry',palette="icefire")

plt.xticks(rotation=90)
plt.xlabel("Industry", fontsize=13)
plt.ylabel("Number of Managers in Industry", fontsize=13)
plt.title("Industries in Survey", fontsize=15)
```

##### Insights:

about the same number of managers across the industry

### Plot for 'salary' column

```python
# create plot for salary column
sns.displot(data = df, x = 'salary', kind = 'hist', bins=30)

plt.xlabel("Salary", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Distribution of Salary Among Managers", fontsize=14)
```

##### Insights:

majority of managers have a salary between 80k to 120k

### Plot for 'industry_experience' column

```python
# create plot for industry experience column

plt.figure(figsize=(12,6))

ax = sns.histplot(data=df, x='industry_experience', bins=20)

plt.xlabel("Industry Experience ", fontsize=13)
plt.ylabel("Number of Managers", fontsize=13)
plt.title("Industry Experience Among Managers", fontsize=16)
```

##### Insights:

10 to 25 years have the majority

### Plot for 'education' column

```python
# create plot for education levels

plt.figure(figsize=(10,10))

data = df['education'].value_counts()

labels = ["Bachelor's Degree", "Master's Degree", "Some College", "High School"]

colors = sns.color_palette('Accent')

plt.title("Manager's Education Levels", fontsize=16)

ax = plt.pie(data, colors = colors, labels = labels, autopct = '%.0f%%')
```

##### Insights:

97% of managers went to college. more then half graduated with a degree and a third got their masters

### Plot for 'gender' column

```python
# create plot for gender column

plt.figure(figsize=(10,10))

data = df['gender'].value_counts()

labels = ['Male', 'Female']

colors = sns.color_palette('rainbow')

ax = plt.pie(data, colors=colors, labels=labels, autopct='%.0f%%')
```

##### Insights: 

half of the managers are male and half are female

```python
# import pearson and spearmanr methods from scipy.stats library
from scipy.stats import pearsonr, spearmanr
```

### What is the correlation between salary and education?

#### Your Hypothesis = The more education you have the higher the salary

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(df['salary'], df['education'])
```

a moderate positive correlation between education level and salary

#### Are these results statistically significant? How do you know?

yes because the pvalue is 0

### What is the correlation between salary and age?

#### Your Hypothesis = the older you are the higher the salary

```python
# calculate correlation and pvalue of salary and age with pearsonr method
pearsonr(df['salary'], df['age'])
```

there is a moderate positive correlation between age and salary

#### Are these results statistically significant? How do you know? 

yes because the pvalue is 0

### What is the correlation between salary and industry experience? 

#### Your Hypothesis = the more industry experience the higher the salary 

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
spearmanr(df['salary'], df['industry_experience'])
```

moderate positive correlation between salary and industry experience

#### Are these results statistically significant? How do you know?

yes because the pvalue is 0

#### Crosstab of average salary grouped by education and industry

```python
# use pd.crosstab() to calculate mean salary in each industry grouped by education level
pd.crosstab(index=df['industry'], columns=df['education'], values=df['salary'], aggfunc=np.mean).round(0)
```

#### Heatmap of crosstab of average salary grouped by education and industry

```python
# save above result to new variable
crosstab = pd.crosstab(index=df['industry'], columns=df['education'], values=df['salary'], aggfunc=np.mean).round(0)
```

```python
# create heatmap with annotated mean salary in each industry and education level
plt.figure(figsize=(16,10))

ax = sns.heatmap(crosstab, annot=True, linewidths=.5, cmap='YlGnBu', fmt='g', xticklabels=['High School', 'Some College', "Bachelor's Degree", "Master's Degree"])

plt.xlabel("Education Level", fontsize=14)
plt.ylabel("Industry", fontsize=14)

plt.title("Manager Salaries Across Industries by Education Level", fontsize=18)
```

##### Insights:

easy to see and understand the education level by industry of the managers
